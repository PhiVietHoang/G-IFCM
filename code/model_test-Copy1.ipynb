```python
U, S = al.method3(R_train, 2, 1.7, 0.45)
```

```python
X = np.dstack((M,N,H))
P = X.shape[0]
y = np.empty(P, dtype = object)
for i in range(P):
    if al.distance_func(X[i], S[0]) < al.distance_func(X[i], S[1]): y[i] = 'Besni'
    else: y[i] = 'Kecimen'

print(np.sum(y == train[:,-1])/P)
```
